```python
import os
import pypandoc

def convert_files_in_folder(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        try:
            if filename.endswith('.mmd'):
                input_full_path = os.path.join(input_folder, filename)

                # Convert MMD to TeX
                tex_filename = filename.replace('.mmd', '.tex')
                tex_full_path = os.path.join(output_folder, tex_filename)
                if not os.path.exists(tex_full_path):
                    pypandoc.convert_file(input_full_path, 'markdown', format='tex', outputfile=tex_full_path)
                    print(f"Converted {filename} to {tex_filename}")

                    # Convert TeX to HTML
                    html_filename = filename.replace('.mmd', '.html')
                    html_full_path = os.path.join(output_folder, html_filename)
                    if not os.path.exists(html_full_path):
                        pypandoc.convert_file(tex_full_path, 'tex', format='html', outputfile=html_full_path)
                        print(f"Converted {tex_filename} to {html_filename}")
        except Exception as e:
            print(f"Error converting {filename}: {e}")

input_folder_path = r'C:\Users\Shadow\Downloads\dev_mmd_annotated'
output_folder_path = r'C:\Users\Shadow\Downloads\dev_html_annotated'
convert_files_in_folder(input_folder_path, output_folder_path)
```
